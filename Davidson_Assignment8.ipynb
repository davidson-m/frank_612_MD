Read the summarize_data() and explain the difference between g and df.

In the summarize_data() function below, g refers to the variable supplied to the function. df is the default dataframe already loaded. This means that if g, the local variable within the function, refers to the same object as df, the "mix" result can end up dividing the same metrics by itself.

```python
import pandas as pd

# the mpg data is used for demonstration
df = pd.read_csv("https://raw.githubusercontent.com/franklin-univ-data-science/data/master/mpg.csv")
df.head()
```

```python
def summarize_data(g):
  return(
    pd.Series({
      "avg_MPG": g.mpg.mean(),
      "min_MPG": g.mpg.min(),
      "max_MPG": g.mpg.max(),
      "sample_count": len(g),
      "mix": g.cylinders.sum()/df.cylinders.sum(),
    })
  )

df.groupby("cylinders").apply(summarize_data)
```

```python
# Load baseball data set from previous assignment:

df2 = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Batting.csv')

df2.head()
```

```python
# Apply summarize_data() function to baseball data set:

def summarize_data(g):
  return(
    pd.Series({
      "avg_BB": g['BB'].mean(),
      "min_BB": g['BB'].min(),
      "max_BB": g['BB'].max(),
      "sample_count": len(g),
      "mix": g['BB'].sum()/df2['BB'].sum(),
    })
  )

df2.groupby('teamID').apply(summarize_data)
```
