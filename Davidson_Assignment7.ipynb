1) Working with Strings

```python
# Clean a column using regular expression methods
# Store the cleaned column into another column and call it "your_col_name_cleaned"

import pandas as pd
import re

df_unclean = pd.read_csv('https://raw.githubusercontent.com/chendaniely/pandas_for_everyone/master/data/scientists.csv')

df_unclean.head()
```

```python
# Select only the year (first four digits) from the Born column, save to new column your_column_name_cleaned

p = '\d\d\d\d'

df_unclean['your_column_name_cleaned'] = df_unclean['Born'].str.split(p)
```

2) Use of .apply()

```python
# Select only numeric columns

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Batting.csv')

df_mod = df.select_dtypes(include='number')

df_mod.info()
```

```python
def mean_func(x):
    return x.mean()

def sum_func(x):
    return x.sum()

def mode_func(x):
    return x.mode()

def median_func(x):
    return x.median()

def range_func(x):
    return x.max() - x.min()

print("Means")
print(df_mod.apply(mean_func))
print("")
print("Sums")
print(df_mod.apply(sum_func))
print("")
print("Modes")
print(df_mod.apply(mode_func))
print("")
print("Medians")
print(df_mod.apply(median_func))
print("")
print("Range")
print(df_mod.apply(range_func))
```
