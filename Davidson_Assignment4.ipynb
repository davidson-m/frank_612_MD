```python
# Scatterplot
# Used to Show the Relationship Between Two Continuous Variables

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Batting.csv')

scatter, ax = plt.subplots()

ax = sns.regplot(x='SO', y='HR', data=df)

ax.set_title('Scatterplot of Strikeouts and Home Runs for Hitters')
ax.set_xlabel('Strikeouts (SO)')
ax.set_ylabel('Home Runs (HR)')

plt.show()
```

```python
# 2D Kernel Density Plot
# Shows concentration of data

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Batting.csv')

kde, ax = plt.subplots()
ax = sns.kdeplot(x=df['SO'], y=df['HR'], shade=True)

ax.set_title('Kernel Density Plot of Strikeouts and Home Runs for Hitters')
ax.set_xlabel('Strikeouts (SO)')
ax.set_ylabel('Home Runs (HR)')

plt.show()
```

```python
# Hexbin joint plot

# Useful to better depict relationship when there are many data points, such as in this data set

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Batting.csv')

hexbin = sns.jointplot(x="SO", y="HR", data=df, kind="hex")

hexbin.set_axis_labels(xlabel='SO', ylabel='HR')

hexbin.fig.suptitle('Hexbin Joint Plot of Strikeouts and Home Runs for Hitters',

                 fontsize=12, y=1.03)

plt.show()
```
